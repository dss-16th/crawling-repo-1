```python
import datetime 
from datetime import datetime

date = 20210302

def input_date(date,**kwargs):
    day = str(date)
    datetimeobject = datetime.strptime(day,'%Y%m%d')
    newformat = datetimeobject.strftime('%Y-%m-%d')
    new = newformat.split("-")
    y = int(new[0])
    m = int(new[1])
    d = int(new[2])


    return y, m, d

 

def is_leap(num):

    if (num % 4) ==0:

        if (num % 400) == 0: 

            return True

        elif (num % 100) == 0: 

            return False

        else: 

            return True

    else: 

        return False

def get_day_name(y, m, d):

    dt = datetime(y, m, d)

    dt = int(dt.strftime("%j"))

    count_day = 0

    for i in range(year):

        if is_leap(i): 

            count_day += 366

        else: 

            count_day += 365

    count = (count_day + dt - 366) %  7 

    #1년 1월 1일은 0일째이지만 컴퓨터에서는 366일째로 계산되기 때문에 366을 빼줌.

    return {0: '일요일', 1: "월요일", 2: "화요일", 3: "수요일",

            4: "목요일", 5: "금요일", 6: "토요일"}[count]


if __name__ == "__main__":

    year, month, day = input_date(date)

    day_name = get_day_name(year, month, day)
    
#     date = date(date)

    print(day_name)


def date(date):
    if day_name == "월요일":
        date
    elif day_name == "화요일":
        date = date-1
    elif day_name == "수요일":
        date = date-2
    elif day_name == "목요일":
        date = date-3
    elif day_name == "금요일":
        date = date-4
    elif day_name == "토요일":
        date = date-5
    return date

date
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page,'lxml')
# entity-title style-scope ytmc-chart-table
title = [each.span.string for each in soup.find_all('div',class_='entity-title style-scope ytmc-chart-table')]
len(title), title
```

```python
# singer = soup.find_all('span','ytmc-artist-name clickable style-scope ytmc-artists-list')
singer = [each.span.string for each in soup.find_all('div',class_='ytmc-artists-list-container style-scope ytmc-artists-list')]
#singer = [[n].string for n in range(0,len(singer))]]
len(singer), singer
```

```python
duration = [each.span.string for each in soup.find_all('div',class_='chart-period style-scope ytmc-chart-table')]
duration 
```

```python
views = [each.span.string for each in soup.find_all('div',class_='views style-scope ytmc-chart-table')]
len(views),views
```
